```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Data_Hydro_Old = pd.read_csv("Concatenated_Old.csv")
Data_Hydro_New = pd.read_csv("Concatenated_New.csv")
Data_Hydro_Old['Run_Time'].hist(bins=80, grid=False, alpha = 0.5, label = 'Fixed grid', color = 'blue', xlabelsize=12, ylabelsize=12)
Data_Hydro_New['Run_Time'].hist(bins=63, grid=False, alpha = 0.5, label = 'Auto grid', color = 'red', xlabelsize=12, ylabelsize=12)
plt.xlabel("Run time (s)", fontsize=15)
plt.ylabel("Counts",fontsize=15)
plt.xlim([0, 520])
plt.ylim([0, 900])
plt.legend(prop={'size': 12})
plt.title('Run time of hydrodynamic simulations', fontsize=16)#, color='red')
plt.savefig('benchmarks.png')
plt.savefig('benchmarks2.pdf')
plt.savefig('benchmarks3.jpeg')
```
